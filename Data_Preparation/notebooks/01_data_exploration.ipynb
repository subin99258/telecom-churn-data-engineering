```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(" ", "_")
              .str.replace(r"[^a-zA-Z0-9_]", "", regex=True)
)
df.columns.tolist()
```

```python
df.info()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(exclude=np.number).columns.tolist()

print("Numeric columns:", numeric_cols)
print("\nCategorical columns:", categorical_cols)
```

```python
df_clean = df.copy()
```

```python
for col in df_clean.select_dtypes(include="object").columns:
    df_clean[col] = df_clean[col].str.strip().str.lower()

df_clean.head()
```

```python
for col in df_clean.select_dtypes(include="object").columns:
    print(f"\n--- {col} ---")
    print(df_clean[col].value_counts(dropna=False))
```

```python
X = df_clean.drop("churn", axis=1)
y = df_clean["churn"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
y = y.map({"no": 0, "yes": 1})
y.head()
```

```python
y.value_counts()
```

```python
numeric_features = ["seniorcitizen", "tenure", "monthlycharges"]
categorical_features = ["gender", "dependents", "phoneservice", "multiplelines", "internetservice", "contract"]

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)
```

```python
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)
X_encoded.head()
```

```python
X_encoded.shape
```

```python
X_encoded.columns.tolist()
```

```python
final_df = X_encoded.copy()
final_df["churn"] = y

final_df.head()
```

```python
final_df.shape
```

```python
final_df.dtypes
```

```python
bool_cols = final_df.select_dtypes(include="bool").columns
final_df[bool_cols] = final_df[bool_cols].astype(int)

final_df.dtypes
```

```python
final_df.to_csv("../data/processed/churn_processed.csv", index=False)
print("Processed dataset saved successfully.")
```
